```python
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import PAIRWISE_DISTANCE_FUNCTIONS
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load("../extracted-embeddings/efficientnetb0_4096_pretrained_train.pickle", allow_pickle=True)
data['embedding'].shape
```

```python
# plt.ion()
# %matplotlib widget
```

```python
OUTPUT_DIM = 2
```

```python
def plot_embed(data, labels):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(data[:,0], data[:,1], data[:,2], c=labels)
    plt.show()
```

```python
tsne = TSNE(n_components=3, init='pca', perplexity=40, verbose=1, metric='cosine')#, method='exact')
```

```python
output = tsne.fit_transform(data['embedding'])
output.shape
```

```python
plot_embed(output, data['target'])
```
